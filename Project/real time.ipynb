```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
import importlib
import time
import scraper as sc

importlib.reload(sc)

web_scraper = sc.scraper()

fig = plt.figure(figsize = (12, 10))
ax_1 = fig.add_subplot(221)
ax_2 = fig.add_subplot(222)
ax_3 = fig.add_subplot(223)
plt.ion()

fig.show()
fig.canvas.draw()

y_1 = []
y_2 = []
y_3 = []
y_4 = []
y_5 = []

for i in range(10):
        ax_1.clear()
        ax_2.clear()
        ax_3.clear()
        #web_scraper.fetch('Bitfinex', y_1)
        web_scraper.fetch('Bittrex', y_2)
        web_scraper.fetch('Gdax', y_3)
        web_scraper.fetch('Bitstamp', y_4)
        web_scraper.fetch('Bithumb', y_5)
        
        #ax_1.plot(y_1, label = 'Bitfinex', color = 'b')
        ax_1.plot(y_2, label = 'Bittrex' if i == 0 else "", color = 'g')
        ax_1.plot(y_3, label = 'Gdax' if i == 0 else "", color = 'r')
        ax_1.plot(y_4, label = 'Bitstamp' if i == 0 else "", color = 'c')
        ax_1.plot(y_5, label = 'Bithumb' if i == 0 else "", color = 'y')
        ax_1.legend(loc="upper right")
        
        max_list = np.maximum.reduce([y_2, y_3, y_4, y_5]).tolist()
        min_list = np.minimum.reduce([y_2, y_3, y_4, y_5]).tolist()
        result = np.subtract(max_list, min_list).tolist()
        ax_2.plot(result)
        
        
        fig.canvas.draw()
        time.sleep(10)
```
